## 锔 Regla de Born
En mec谩nica cu谩ntica, la **Regla de Born** establece c贸mo se relaciona un estado cu谩ntico con las probabilidades de obtener un resultado al medirlo.  

Si un sistema est谩 en el estado puro
$$
|\psi\rangle = \sum_i a_i |i\rangle
$$

donde $|i\rangle\>$ es la **base computacional**, entonces la probabilidad de obtener el resultado $i$ al medir es:

$$
p(i) = |a_i|^2
$$

- $a_i$ son las **amplitudes complejas** del estado.  
- $|a_i|^2$ es siempre real y no negativo.  
- La suma de probabilidades debe cumplir:  
$$
\sum_i |a_i|^2 = 1
$$

En otras palabras, la medici贸n proyecta el estado sobre uno de los vectores base con una probabilidad dada por el cuadrado del m贸dulo de su amplitud.  

---

##  Ejemplo num茅rico de la Regla de Born

Consideremos un qubit en el estado:

$$
|\psi\rangle = \sqrt{\tfrac{3}{4}}|0\rangle + \sqrt{\tfrac{1}{4}}|1\rangle
$$

- Amplitudes:  
  $$
  a_0 = \sqrt{\tfrac{3}{4}}, \quad a_1 = \sqrt{\tfrac{1}{4}}
  $$

- Probabilidades por la **Regla de Born**:  
  $$
  p(0) = |a_0|^2 = 0.75, \quad p(1) = |a_1|^2 = 0.25
  $$

Esto significa que al medir muchas veces este estado:  
- Aproximadamente el **75%** de los resultados ser谩n $|0\rangle$,  
- y el **25%** de los resultados ser谩n $|1\rangle$.

## 2. `BalancedState`
Construye un **estado balanceado** (igual superposici贸n) para $n$ qubits.  

- Para un solo qubit:  
  $$
  |\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)
  $$
- Para varios qubits, se usa el producto tensorial (Kronecker $\otimes$) repetidamente.  
Esto genera un estado de superposici贸n uniforme sobre $2^n$ estados.

Ejemplo:  
Para 2 qubits:  
$$
|\psi\rangle = \frac{1}{2}(|00\rangle + |01\rangle + |10\rangle + |11\rangle)
$$

>C贸digo en Python
```python
import numpy as np

def BalancedState(numberQbits):
  initial = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
  if numberQbits > 1:
    psi_h = intial
    for _ in range(numberQbits - 1):
      v = np.kron(initial, psi_h)
    else:
      v=initial
  
  return v
