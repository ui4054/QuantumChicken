# üìä Visualizaci√≥n de vectores probabil√≠sticos con `plt.bar`

## üîπ Objetivo
Representar un **vector de probabilidad** en un **diagrama de barras**.  
Ejemplo: una moneda cargada con probabilidades:

$$
\vec{p} = 
\begin{bmatrix}
0.3 \\ 
0.7
\end{bmatrix}
$$

donde $p_s = 0.3$ (sello) y $p_c = 0.7$ (cara).  

---

## üîπ Sintaxis de `plt.bar`

La funci√≥n principal es `plt.bar()`. Cada argumento permite configurar distintos aspectos del gr√°fico.

### Par√°metros principales

- **`x`**  
  Posiciones o etiquetas en el eje horizontal.  
  - Ejemplo: `["Sello", "Cara"]`.

- **`height`**  
  Alturas de cada barra.  
  - En este caso: `[0.3, 0.7]` (las probabilidades).

### Par√°metros opcionales de personalizaci√≥n

- **`width=0.8`**  
  Grosor de las barras (por defecto 0.8).  

- **`bottom=None`**  
  Punto en el eje Y desde donde arranca cada barra.  
  - Si no se especifica, arranca en `0`.  

- **`align='center'`**  
  Controla c√≥mo se alinean las barras respecto a `x`.  
  - `'center'`: centradas en el valor.  
  - `'edge'`: alineadas por el borde.  

- **`color=None`**  
  Define los colores de las barras.  
  - Puede ser un solo color (`"blue"`) o una lista (`["skyblue", "orange"]`).  

- **`edgecolor=None`**  
  Color de los bordes de cada barra.  

- **`linewidth=None`**  
  Grosor del borde de cada barra.  

- **`tick_label=None`**  
  Etiquetas de las posiciones `x` (alternativa a pasarlas directamente como lista).  

---

## üîπ Resumen sint√°ctico

```python
plt.bar(x, height, width=0.8, bottom=None, align='center', color=None, edgecolor=None, linewidth=None, tick_label=None)


````python
import numpy as np
import matplotlib.pyplot as plt

# Vector de probabilidad de la moneda [Sello, Cara]
p = np.array([0.3, 0.7])

# Etiquetas para el eje X
etiquetas = ["Sello", "Cara"]

# Crear gr√°fico de barras
plt.bar(
    x=etiquetas,      # categor√≠as en el eje X
    height=p,         # altura de cada barra (las probabilidades)
    width=0.6,        # ancho de cada barra (m√°s angostas que el valor por defecto)
    bottom=0,         # las barras parten desde 0 en el eje Y
    align='center',   # cada barra queda centrada en su categor√≠a
    color=["skyblue", "orange"],  # colores personalizados para cada barra
    edgecolor="black", # color del borde de las barras
    linewidth=1.2      # grosor del borde
)

# Agregar t√≠tulo y etiquetas
plt.title("Probabilidades de la moneda")
plt.ylabel("Probabilidad")
plt.xlabel("Resultado")

# Mostrar el gr√°fico
plt.show()
````

# ü•ß Visualizaci√≥n de vectores probabil√≠sticos con `plt.pie`

## üîπ Objetivo
Representar un **vector de probabilidad** como un **gr√°fico circular** (gr√°fico de torta o "pie chart").  
Este tipo de gr√°fico permite **comparar proporciones** de manera visual.

Supongamos la misma moneda cargada con probabilidades:

$$
\vec{p} =
\begin{bmatrix}
0.3 \\
0.7
\end{bmatrix}
=
\begin{bmatrix}
p_s \\
p_c
\end{bmatrix}
$$

donde $p_s = 0.3$ (sello) y $p_c = 0.7$ (cara).  

---

## üîπ Sintaxis de `plt.pie`

La funci√≥n principal es `plt.pie()`.  

### Par√°metros principales

- **`x`**  
  Lista de valores que se van a representar en proporciones.  
  - Ejemplo: `[0.3, 0.7]`.

- **`labels`** *(opcional)*  
  Etiquetas para cada porci√≥n.  
  - Ejemplo: `["Sello", "Cara"]`.

- **`colors`** *(opcional)*  
  Colores asignados a cada sector.  
  - Ejemplo: `["skyblue", "orange"]`.

- **`autopct`** *(opcional)*  
  Formato de texto que muestra el **porcentaje** en cada sector.  
  - Ejemplo: `"%.1f%%"` ‚Üí mostrar√° un n√∫mero con un decimal y el s√≠mbolo `%`.  

### Par√°metros de estilo

- **`startangle=0`**  
  √Ångulo inicial en grados desde donde comienza el gr√°fico.  
  - Ejemplo: `90` ‚Üí empieza desde la parte superior.

- **`counterclock=True`**  
  Define si las porciones se dibujan en sentido antihorario (`True`) o horario (`False`).  

- **`shadow=False`**  
  Si se activa (`True`), dibuja una sombra debajo del gr√°fico.  

---

## üîπ Resumen sint√°ctico

```python
plt.pie(x, labels=None, colors=None, autopct=None, startangle=0, counterclock=True, shadow=False)


````python
import numpy as np
import matplotlib.pyplot as plt

# Vector de probabilidad de la moneda [Sello, Cara]
p = np.array([0.3, 0.7])

# Etiquetas para cada sector del gr√°fico
etiquetas = ["Sello", "Cara"]

# Colores para diferenciar cada sector
colores = ["skyblue", "orange"]

# Crear gr√°fico de pastel
plt.pie(
    x=p,                  # valores (probabilidades)
    labels=etiquetas,     # etiquetas de cada sector
    colors=colores,       # colores personalizados
    autopct="%.1f%%",     # mostrar porcentajes con 1 decimal
    startangle=90,        # comenzar desde la parte superior
    counterclock=True,    # dibujar en sentido antihorario
    shadow=True           # activar sombra debajo del gr√°fico
)

# Agregar t√≠tulo
plt.title("Distribuci√≥n de probabilidades de la moneda")

# Mostrar gr√°fico
plt.show()
````

# üñºÔ∏è Uso de `plt.subplots` en Matplotlib

## üîπ Objetivo  
`plt.subplots` permite **crear una figura con m√∫ltiples ejes (subplots) organizados en filas y columnas**.  
Esto es √∫til para:  

- Comparar diferentes gr√°ficos lado a lado.  
- Visualizar distintos aspectos del mismo conjunto de datos.  
- Mantener una presentaci√≥n limpia y uniforme.

---

## üîπ Conceptos clave

1. **Figura (`fig`)**  
   - Representa el **contenedor principal** de todos los subplots.  
   - Se puede ajustar su tama√±o con `figsize=(ancho, alto)`.

2. **Ejes (`axs`)**  
   - Son los **subplots individuales** dentro de la figura.  
   - Pueden ser un **array** (si hay m√°s de uno) o un **objeto √∫nico** (si solo hay un subplot).  
   - Cada `ax` se puede usar como un eje independiente para graficar (`ax.plot(...)`, `ax.imshow(...)`, etc.).

3. **Distribuci√≥n de subplots**  
   - Definida por **filas y columnas**: `plt.subplots(nrows, ncols)`.  
   - Ejemplo: `plt.subplots(2, 3)` ‚Üí 2 filas y 3 columnas (6 subplots).  
   - `axs` tendr√° la forma de un **array 2D**: `axs[fila, columna]`.

---

## üîπ Pasos did√°cticos para usar `plt.subplots`

1. **Crear la figura y los ejes**  
   - `fig, axs = plt.subplots(1, 3, figsize=(12, 4))`  
   - Significa **1 fila y 3 columnas**, cada subplot se guarda en `axs[0]`, `axs[1]`, `axs[2]`.

2. **Graficar en cada subplot**  
   - Cada eje es independiente:  
     ```text
     axs[0].imshow(...)  # Primer subplot
     axs[1].imshow(...)  # Segundo subplot
     axs[2].imshow(...)  # Tercer subplot
     ```
   - Se pueden asignar t√≠tulos independientes a cada subplot: `axs[i].set_title("...")`.

3. **Agregar colorbars u otras anotaciones**  
   - Se pueden vincular a un subplot espec√≠fico:  
     `plt.colorbar(im, ax=axs[i])`.

4. **Ajustar espaciado**  
   - `plt.tight_layout()` para que los subplots no se superpongan y se vea limpio.

5. **Mostrar la figura**  
   - `plt.show()` renderiza todos los subplots simult√°neamente.

---

## üîπ Notas pedag√≥gicas

- Cuando se usan **muchos subplots**, conviene que la figura sea m√°s ancha o m√°s alta (`figsize`) para que no queden apretados.  
- Si hay un solo subplot, `axs` puede ser un objeto √∫nico en lugar de un array.  
- Se pueden usar combinaciones de `plt.subplot()` (una sola celda) para gr√°ficos r√°pidos, pero `plt.subplots()` es m√°s flexible para figuras complejas.

---

## üîπ Sintaxis de `plt.subplots` (al final)

```python
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(ancho, alto))


# üå°Ô∏è Visualizaci√≥n de vectores probabil√≠sticos con `plt.imshow`

## üîπ Objetivo
Un **heatmap (mapa de calor)** es una representaci√≥n donde los valores num√©ricos se muestran como **colores**.  
Esto permite visualizar la magnitud de las probabilidades de un vector de forma m√°s intuitiva.

Por ejemplo, para la moneda con probabilidades:

$$
\vec{p} =
\begin{bmatrix}
0.3 \\
0.7
\end{bmatrix}
$$

podemos representar estos valores en una escala de colores.

---

## üîπ Sintaxis de `plt.imshow`

La funci√≥n es:

### Par√°metros principales

- **`X`**  
  Matriz (2D) o vector convertido en matriz, cuyos valores ser√°n representados como colores.  
  - Ejemplo: `[[0.3, 0.7]]` para visualizar un vector fila.

- **`cmap`** *(opcional)*  
  Mapa de colores usado para la escala.  
  Existen distintos esquemas que conviene elegir seg√∫n el tipo de datos:

  - `"viridis"`: mapa perceptualmente uniforme, pasa del morado al amarillo. Es el **mapa por defecto recomendado** en Matplotlib. Ideal para datos continuos.  
  - `"plasma"`: colores brillantes, del p√∫rpura al amarillo intenso. Bueno para resaltar contrastes.  
  - `"inferno"`: del negro al naranja y amarillo, √∫til en datos que representan calor o energ√≠a.  
  - `"magma"`: del negro al rosa p√°lido, m√°s suave que *inferno*.  
  - `"cividis"`: dise√±ado para ser accesible a personas con daltonismo (buena opci√≥n inclusiva).  
  - `"hot"`: pasa de negro ‚Üí rojo ‚Üí amarillo ‚Üí blanco, t√≠pico en mapas t√©rmicos.  
  - `"coolwarm"`: va de azul a rojo, √∫til para datos centrados en un valor medio (positivo vs negativo).  
  - `"Greys"`: escala de grises, √∫til para resaltar intensidad sin color.

- **`aspect`** *(opcional)*  
  Controla la relaci√≥n ancho/alto del gr√°fico.  
  - `"auto"` ajusta autom√°ticamente.  
  - `"equal"` fuerza que cada celda sea cuadrada.

- **`interpolation`** *(opcional)*  
  M√©todo de suavizado entre valores.  
  - `"nearest"` mantiene bordes n√≠tidos.  
  - `"bilinear"` suaviza la transici√≥n de colores.

---

## üîπ Escala de colores

Para agregar la barra lateral que muestra qu√© color corresponde a qu√© valor, usamos:

plt.colorbar()

## üîπ Sintaxis

```python
plt.imshow(X, cmap=None, aspect=None, interpolation=None)



````python
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 1. Definir el vector de probabilidades
# ==============================
# Supongamos que lanzamos una moneda "no perfectamente justa".
# La probabilidad de Sello = 0.3
# La probabilidad de Cara  = 0.7
# Lo representamos como un vector:
# p = [P(Sello), P(Cara)]
p = np.array([0.3, 0.7])

# ==============================
# 2. Convertir el vector a una "matriz fila"
# ==============================
# imshow espera una matriz 2D para poder graficar.
# reshape(1, -1) convierte [0.3, 0.7] ‚Üí [[0.3, 0.7]]
# El "1" indica una fila, y "-1" deja que numpy infiera el n√∫mero de columnas.
p_matriz = p.reshape(1, -1)

# ==============================
# 3. Crear la figura con varios subplots
# ==============================
# Creamos una figura con 1 fila y 3 columnas de ejes (axs es un arreglo con 3 objetos Axes).
# figsize=(12,4) define el tama√±o total de la figura.
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# ==============================
# 4. Primer heatmap con el mapa de colores "viridis"
# ==============================
# Dibujamos la matriz como un heatmap en el primer eje (axs[0]).
# cmap="viridis" usa un degradado verde-azul-amarillo.
im1 = axs[0].imshow(p_matriz, cmap="viridis", aspect="auto")
axs[0].set_title("cmap = viridis")       # T√≠tulo del subplot
plt.colorbar(im1, ax=axs[0])             # Barra de color asociada al heatmap

# ==============================
# 5. Segundo heatmap con el mapa de colores "inferno"
# ==============================
# "inferno" usa un degradado oscuro ‚Üí rojo ‚Üí amarillo.
im2 = axs[1].imshow(p_matriz, cmap="inferno", aspect="auto")
axs[1].set_title("cmap = inferno")
plt.colorbar(im2, ax=axs[1])

# ==============================
# 6. Tercer heatmap con el mapa de colores "cividis"
# ==============================
# "cividis" fue dise√±ado para ser perceptible por personas con daltonismo.
im3 = axs[2].imshow(p_matriz, cmap="cividis", aspect="auto")
axs[2].set_title("cmap = cividis")
plt.colorbar(im3, ax=axs[2])

# ==============================
# 7. Ajustar el espaciado entre subplots
# ==============================
# Evita que los t√≠tulos o etiquetas se sobrepongan entre s√≠.
plt.tight_layout()

# ==============================
# 8. Mostrar la figura completa
# ==============================
plt.show()
````

````python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 1. Definir la matriz de transici√≥n
# ==============================
# La moneda tiene dos estados: Sello (0) y Cara (1).
# Cada fila = estado actual.
# Cada columna = estado siguiente.
# La probabilidad de pasar de un estado a otro es 0.5 en todos los casos.
matriz_transicion = np.array([
    [0.1, 0.9],   # Fila 0: Si est√° en Sello ‚Üí 50% Sello, 50% Cara
    [0.9, 0.2]    # Fila 1: Si est√° en Cara  ‚Üí 50% Sello, 50% Cara
])

# ==============================
# 2. Crear la figura y el eje
# ==============================
# plt.subplots devuelve una tupla con dos objetos:
#   - El "lienzo" (Figure), donde se colocan los ejes.
#   - El "sistema de coordenadas" (Axes), donde se dibuja.
figura, eje = plt.subplots(figsize=(5, 4))  
# figsize=(5,4) define el tama√±o de la figura en pulgadas (ancho x alto).

# ==============================
# 3. Dibujar la matriz como imagen (heatmap)
# ==============================
# imshow transforma la matriz num√©rica en una representaci√≥n gr√°fica de colores.
imagen = eje.imshow(
    matriz_transicion,  # la matriz que queremos visualizar
    cmap="Blues",       # esquema de colores
    aspect="equal"      # mantiene las celdas cuadradas
)

# ==============================
# 4. Configurar los ejes
# ==============================
# Definimos las etiquetas que representan los estados de la moneda.
etiquetas_estados = ["Sello", "Cara"]

# Posiciones de los ticks (coinciden con los √≠ndices de la matriz: 0 y 1).
eje.set_xticks([0, 1])  
eje.set_yticks([0, 1])  

# Asignamos los nombres de estados a esas posiciones.
eje.set_xticklabels(etiquetas_estados)  
eje.set_yticklabels(etiquetas_estados)  

# ==============================
# 5. T√≠tulos y r√≥tulos
# ==============================
eje.set_title("Matriz de transici√≥n de la moneda")  # t√≠tulo general
eje.set_xlabel("Estado siguiente")                  # eje X = columnas
eje.set_ylabel("Estado actual")                     # eje Y = filas

# ==============================
# 6. Barra de colores (leyenda)
# ==============================
# La barra de color explica c√≥mo los valores de la matriz se traducen en colores.
plt.colorbar(imagen, ax=eje)

# ==============================
# 7. Mostrar el gr√°fico
# ==============================
plt.show()
````

# üìä Visualizaci√≥n de frecuencias con `plt.hist`

## üîπ Objetivo
Un **histograma** es una representaci√≥n gr√°fica que muestra la **frecuencia** con la que ocurren ciertos valores dentro de un conjunto de datos.  
En lugar de mostrar cada valor individualmente, los agrupa en **intervalos (bins)** y dibuja barras para representar cu√°ntos datos caen en cada uno.  

Por ejemplo, si simulamos **50 sorteos de loter√≠a** con n√∫meros de 4 cifras (0000‚Äì9999), podemos analizar la **frecuencia de aparici√≥n de las cifras finales (0‚Äì9)** en esos sorteos.  

De esta manera, tenemos un vector de frecuencias que describe la distribuci√≥n emp√≠rica de los d√≠gitos:  

$$
\vec{f} =
\begin{bmatrix}
f(0) & f(1) & f(2) & \dots & f(9)
\end{bmatrix}
$$

donde $f(i)$ es la frecuencia del d√≠gito $i$ en las unidades de los sorteos.

---

## üîπ Par√°metros principales

- **`x`**  
  Lista, array o conjunto de datos num√©ricos que se desean analizar.  
  - Ejemplo: las √∫ltimas cifras de los 50 n√∫meros sorteados.  

- **`bins`**  
  N√∫mero de intervalos (o bordes de clase) en los que se agrupan los datos.  
  - Ejemplo: `bins=10` para separar los n√∫meros en 10 grupos (0‚Äì9).  

- **`density`** *(opcional)*  
  - `False` (por defecto): el histograma muestra frecuencias absolutas.  
  - `True`: el histograma muestra frecuencias relativas (probabilidades), normalizando el √°rea total a 1.  

- **`color`** *(opcional)*  
  Define el color de las barras. Ejemplo: `"skyblue"`, `"lightgreen"`, `"salmon"`.  

- **`edgecolor`** *(opcional)*  
  Define el color de los bordes de las barras. √ötil para mejorar la visibilidad.  

- **`alpha`** *(opcional)*  
  Controla la transparencia de las barras (`0` totalmente transparente, `1` totalmente opaco).  

---

## üîπ Sintaxis de `plt.hist`

```python
plt.hist(x, bins=None, density=False, color=None, edgecolor=None, alpha=None)


````python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# üé≤ 1. Generar datos simulados de la loter√≠a
# ----------------------------------------------------
# Vamos a simular 50 sorteos de un n√∫mero de 4 cifras (0000 a 9999).
# Usamos randint para obtener n√∫meros enteros en ese rango.
sorteos = np.random.randint(0, 10000, size=50)

# Extraemos la √∫ltima cifra (unidades) de cada n√∫mero.
# Esto se hace con el operador m√≥dulo % 10.
ultimas_cifras = sorteos % 10

# ----------------------------------------------------
# üìä 2. Graficar el histograma
# ----------------------------------------------------
plt.figure(figsize=(8, 5))  # Definimos tama√±o de la figura

# Creamos el histograma
plt.hist(
    ultimas_cifras,     # Datos a graficar
    bins=10,            # 10 intervalos (0 al 9)
    color="skyblue",    # Color de las barras
    edgecolor="black",  # Color de los bordes de las barras
    alpha=0.7           # Transparencia de las barras
)

# ----------------------------------------------------
# üìù 3. Personalizar el gr√°fico
# ----------------------------------------------------
plt.title("Distribuci√≥n de las √∫ltimas cifras en 50 sorteos de loter√≠a")
plt.xlabel("Cifra final (0‚Äì9)")
plt.ylabel("Frecuencia")
plt.xticks(range(10))  # Etiquetas del eje X (0 a 9)

# Mostrar gr√°fico
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# üé≤ 1. Simulaci√≥n de los sorteos
# ----------------------------------------------------
# Generamos 50 n√∫meros enteros entre 0 y 9999 (inclusive).
# Esto simula 50 sorteos de loter√≠a de 4 cifras.
sorteos = np.random.randint(0, 10000, size=50)

# Convertimos cada n√∫mero entero en un string de 4 d√≠gitos.
# Usamos f"{num:04d}" para que siempre tenga 4 posiciones:
#  - el n√∫mero 42 se convierte en "0042"
#  - el n√∫mero 7 en "0007"
# Esto facilita extraer cada cifra por separado.
sorteos_str = [f"{num:04d}" for num in sorteos]

# Extraemos cada d√≠gito por posici√≥n:
#  - miles: car√°cter en posici√≥n [0] del string
#  - centenas: car√°cter en posici√≥n [1]
#  - decenas: car√°cter en posici√≥n [2]
#  - unidades: car√°cter en posici√≥n [3]
# Convertimos cada car√°cter a entero con int()
miles    = [int(num[0]) for num in sorteos_str]
centenas = [int(num[1]) for num in sorteos_str]
decenas  = [int(num[2]) for num in sorteos_str]
unidades = [int(num[3]) for num in sorteos_str]

# ----------------------------------------------------
# üìä 2. Cuatro histogramas separados (uno por d√≠gito)
# ----------------------------------------------------
# Creamos una figura con 2 filas y 2 columnas de subgr√°ficos (axs ser√° una matriz 2x2)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# --- Histograma de los miles ---
axs[0, 0].hist(miles, bins=10, color="skyblue", edgecolor="black")
# bins=10 -> intervalos para cada d√≠gito (0 al 9)
# color -> color de relleno de las barras
# edgecolor -> borde negro para mayor contraste
axs[0, 0].set_title("Distribuci√≥n de los miles")
axs[0, 0].set_xticks(range(10))  # Mostramos los d√≠gitos 0‚Äì9 en el eje X

# --- Histograma de las centenas ---
axs[0, 1].hist(centenas, bins=10, color="salmon", edgecolor="black")
axs[0, 1].set_title("Distribuci√≥n de las centenas")
axs[0, 1].set_xticks(range(10))

# --- Histograma de las decenas ---
axs[1, 0].hist(decenas, bins=10, color="lightgreen", edgecolor="black")
axs[1, 0].set_title("Distribuci√≥n de las decenas")
axs[1, 0].set_xticks(range(10))

# --- Histograma de las unidades ---
axs[1, 1].hist(unidades, bins=10, color="orange", edgecolor="black")
axs[1, 1].set_title("Distribuci√≥n de las unidades")
axs[1, 1].set_xticks(range(10))

# T√≠tulo general de la figura
plt.suptitle("Distribuci√≥n de cada d√≠gito en los 50 sorteos", fontsize=14)

# Ajustamos los espacios entre subplots para que no se encimen
plt.tight_layout()

# Mostramos la figura en pantalla
plt.show()

# ----------------------------------------------------
# üìä 3. Gr√°fico combinado con 4 series (barras lado a lado)
# ----------------------------------------------------
plt.figure(figsize=(10, 6))  # Figura nueva

# Calculamos las frecuencias absolutas de cada d√≠gito en cada posici√≥n
# np.histogram devuelve dos cosas:
#   - counts: cu√°ntos elementos caen en cada bin (intervalos)
#   - edges: los bordes de los intervalos (en este conteo no se usan)
fre_miles, _    = np.histogram(miles, bins=10, range=(0, 10))
fre_centenas, _ = np.histogram(centenas, bins=10, range=(0, 10))
fre_decenas, _  = np.histogram(decenas, bins=10, range=(0, 10))
fre_unidades, _ = np.histogram(unidades, bins=10, range=(0, 10))

# Creamos un vector con las posiciones del eje X: 0, 1, 2, ..., 9
x = np.arange(10)

# Definimos el ancho de cada barra
width = 0.2

# Dibujamos las 4 series de barras:
#   - Para que no se encimen, desplazamos cada grupo en el eje X
#     (ejemplo: miles va en x-1.5*width, centenas en x-0.5*width, etc.)
plt.bar(x - 1.5*width, fre_miles, width, label="Miles", color="skyblue", edgecolor="black")
plt.bar(x - 0.5*width, fre_centenas, width, label="Centenas", color="salmon", edgecolor="black")
plt.bar(x + 0.5*width, fre_decenas, width, label="Decenas", color="lightgreen", edgecolor="black")
plt.bar(x + 1.5*width, fre_unidades, width, label="Unidades", color="orange", edgecolor="black")

# Personalizamos el gr√°fico
plt.xticks(x, range(10))  # Etiquetas 0‚Äì9 en eje X
plt.xlabel("D√≠gito (0‚Äì9)")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n comparativa de cada posici√≥n en los sorteos")
plt.legend()  # Muestra la leyenda con "Miles", "Centenas", etc.

# Mostramos en pantalla
plt.show()
````

# üèπ La funci√≥n `plt.quiver` en Matplotlib

## üîπ Descripci√≥n
`plt.quiver` permite representar **vectores como flechas** en un plano cartesiano.  
Es muy utilizada para mostrar campos vectoriales, direcciones y magnitudes.

---

## üîπ Par√°metros principales

- **`X`, `Y`**  
  Coordenadas donde comienzan las flechas.  
  Se pueden dar como escalares, listas o matrices.

- **`U`, `V`**  
  Componentes de cada flecha:  
  - `U` ‚Üí desplazamiento en el eje **X** (horizontal).  
  - `V` ‚Üí desplazamiento en el eje **Y** (vertical).  

- **`angles`** *(opcional)*  
  Define c√≥mo se interpretan los √°ngulos:  
  - `"xy"`: coordenadas cartesianas (lo m√°s com√∫n).  
  - `"uv"`: interpreta directamente `U, V` como direcci√≥n.  

- **`scale`** *(opcional)*  
  Factor de escala para la longitud de las flechas.  
  Si es `1`, las flechas mantienen las magnitudes originales.  

- **`scale_units`** *(opcional)*  
  Define las unidades usadas para la escala: `"xy"`, `"width"`, `"height"`, etc.  

- **`color`** *(opcional)*  
  Color de las flechas. Puede ser un solo valor o una lista.  

- **`width`** *(opcional)*  
  Grosor de las flechas en la figura.  

---

# üé® Etiquetas, t√≠tulos y ejes en Matplotlib (sin morir en el intento)

Cuando usamos gr√°ficos, no basta con poner flechas o colores:  
necesitamos **explicar** qu√© est√° pasando.  
Aqu√≠ entran en juego las **etiquetas y adornos visuales**.

---

## üîπ T√≠tulo y etiquetas de ejes

- **`plt.title("...")`**  
  Agrega un t√≠tulo al gr√°fico.  
  Ejemplo: `"Amplitudes complejas de la moneda"`.

- **`plt.xlabel("...")`**  
  Define el nombre del eje **X**.  

- **`plt.ylabel("...")`**  
  Define el nombre del eje **Y**.  

---

## üîπ Grilla

- **`plt.grid(True)`**  
  Muestra una rejilla en el fondo del gr√°fico.  
  Esto facilita visualizar posiciones y magnitudes.  

---

## üîπ L√≠neas de referencia en los ejes

- **`plt.axhline(0, color="black", linewidth=0.5)`**  
  Dibuja una l√≠nea horizontal en el valor `0`.  
  Es √∫til para remarcar el **eje X**.

- **`plt.axvline(0, color="black", linewidth=0.5)`**  
  Dibuja una l√≠nea vertical en el valor `0`.  
  Es √∫til para remarcar el **eje Y**.

---

## üîπ L√≠mites de la ventana

- **`plt.xlim(a, b)`**  
  Fija los l√≠mites de visualizaci√≥n del eje **X** entre `a` y `b`.  

- **`plt.ylim(c, d)`**  
  Fija los l√≠mites de visualizaci√≥n del eje **Y** entre `c` y `d`.

````python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# 1. Definici√≥n del estado cu√°ntico de la moneda
# ----------------------------------------------------
# Amplitudes complejas:
# Definimos un estado |œà> = (1/‚àö2) * [1, i]
# Œ± = 1/‚àö2   (para "Sello")
# Œ≤ = i/‚àö2   (para "Cara")
alpha = 1/np.sqrt(2) * 1       # amplitud para "Sello"
beta  = 1/np.sqrt(2) * 1j      # amplitud para "Cara"

# Construimos el vector de estado como array de NumPy con NOMBRE PSI
psi = np.array([alpha, beta])

# ----------------------------------------------------
# 2. C√°lculo de probabilidades (m√≥dulo al cuadrado)
# ----------------------------------------------------
# Regla cu√°ntica: P(x) = |amplitud|^2
# Esto convierte amplitudes complejas en probabilidades reales
probs = np.abs(psi)**2  # ‚Üí [|Œ±|^2, |Œ≤|^2] = [0.5, 0.5]

# ----------------------------------------------------
# 3. Crear figura con subplots
# ----------------------------------------------------
# Creamos una figura de 10 x 4 pulgadas
plt.figure(figsize=(10,4))

# SUBPLOT 1 (izquierda)
# plt.subplot(filas, columnas, √≠ndice)
# (1, 2, 1) significa:
#   - 1 fila
#   - 2 columnas
#   - este es el gr√°fico N¬∞ 1 (el de la izquierda)
plt.subplot(1, 2, 1)
plt.bar(["Sello", "Cara"], probs, 
        color=["skyblue", "lightcoral"], 
        edgecolor="black")
plt.title("Probabilidades $|\\alpha|^2$, $|\\beta|^2$")
plt.ylabel("Probabilidad")

# SUBPLOT 2 (derecha)
# (1, 2, 2) significa:
#   - 1 fila
#   - 2 columnas
#   - este es el gr√°fico N¬∞ 2 (el de la derecha)
plt.subplot(1, 2, 2)


# ----------------------------------------------------
# Dibujamos el vector de "Sello" (√≠ndice 0 del estado)
# ----------------------------------------------------
amp_sello = psi[0]   # amplitud compleja para "Sello"
plt.quiver(
    0, 0,                # origen (0,0)
    amp_sello.real,      # componente en el eje X
    amp_sello.imag,      # componente en el eje Y
    angles="xy", scale_units="xy", scale=2,
    color="blue", width=0.01
)
plt.text(
    amp_sello.real + 0.05, amp_sello.imag + 0.05,
    f"Sello\n{amp_sello.real}", fontsize=12
)

# ----------------------------------------------------
# Dibujamos el vector de "Cara" (√≠ndice 1 del estado)
# ----------------------------------------------------
amp_cara = psi[1]    # amplitud compleja para "Cara"
plt.quiver(
    0, 0, 
    amp_cara.real,
    amp_cara.imag,
    angles="xy", scale_units="xy", scale=2,
    color="red", width=0.01
)
plt.text(
    amp_cara.real + 0.05, amp_cara.imag + 0.05,
     f"Cara\n{amp_sello.real}", fontsize=12
)

# ----------------------------------------------------
# Etiquetas y t√≠tulo
# ----------------------------------------------------
plt.title("Amplitudes complejas de la moneda")
plt.xlabel("Parte real")
plt.ylabel("Parte imaginaria")
plt.grid(True)



# Dibujamos ejes X e Y
plt.axhline(0, color="black", linewidth=0.5) #
plt.axvline(0, color="black", linewidth=0.5)
plt.title("Amplitudes complejas en el plano (Re, Im)")
plt.xlabel("Parte real")
plt.ylabel("Parte imaginaria")
plt.grid(True)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
#plt.axis("equal")  # escala igual en X e Y para no deformar los vectores

# ----------------------------------------------------
# 4. Mostrar gr√°fico final
# ----------------------------------------------------
plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definici√≥n de vectores de la base can√≥nica en R^3
e1 = np.array([1, 0, 0])
e2 = np.array([0, 1, 0])
e3 = np.array([0, 0, 1])

# Crear figura 3D
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar los vectores
ax.quiver(0, 0, 0, *e1, color='r', label=r'e_1', linewidth=2)
ax.quiver(0, 0, 0, *e2, color='g', label=r'e_2', linewidth=2)
ax.quiver(0, 0, 0, *e3, color='b', label=r'e_3', linewidth=2)

# Configuraci√≥n de los ejes
ax.set_xlim([-0.5, 1.2])
ax.set_ylim([-0.5, 1.2])
ax.set_zlim([-0.5, 1.2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# T√≠tulo y leyenda
ax.set_title("Base ortonormal en $\\mathbb{R}^3$")
ax.legend()

plt.show()
````

````python
# üì¶ Importamos librer√≠as necesarias
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Permite gr√°ficos en 3D

# ------------------------------------------------------
# Definici√≥n de los vectores de la base can√≥nica en R^3
# ------------------------------------------------------
# Cada vector apunta en la direcci√≥n de un eje cartesiano:
e1 = np.array([1, 0, 0])   # Vector unitario a lo largo del eje X
e2 = np.array([0, 1, 0])   # Vector unitario a lo largo del eje Y
e3 = np.array([0, 0, 1])   # Vector unitario a lo largo del eje Z

# ------------------------------------------------------
# Crear figura 3D
# ------------------------------------------------------
fig = plt.figure(figsize=(8,8))  # Hacemos la figura un poco m√°s grande
ax = fig.add_subplot(111, projection='3d')

# ------------------------------------------------------
# Dibujar vectores de la base ortonormal
# ------------------------------------------------------
# Cada vector arranca en el origen (0,0,0) y apunta en la direcci√≥n de su eje
ax.quiver(0, 0, 0, *e1, color='r', label=r'$e_1$', linewidth=2)
ax.quiver(0, 0, 0, *e2, color='g', label=r'$e_2$', linewidth=2)
ax.quiver(0, 0, 0, *e3, color='b', label=r'$e_3$', linewidth=2)

# ------------------------------------------------------
# Dibujar planos coordenados XY, XZ y YZ
# ------------------------------------------------------
# Usamos "meshgrid" para crear una rejilla de puntos que luego se pinta como superficie

# Rango de los ejes
xx = np.linspace(0, 1, 10)
yy = np.linspace(0, 1, 10)
XX, YY = np.meshgrid(xx, yy)

# Plano XY (Z=0)
ZZ = np.zeros_like(XX)
ax.plot_surface(XX, YY, ZZ, alpha=0.2, color='gray')  # Plano Transparente

# Plano XZ (Y=0)
YY = np.zeros_like(XX)
ZZ = np.linspace(0, 1, 10)
XX, ZZ = np.meshgrid(xx, ZZ)
ax.plot_surface(XX, YY, ZZ, alpha=0.2, color='orange')

# Plano YZ (X=0)
XX = np.zeros_like(ZZ)
YY, ZZ = np.meshgrid(yy, np.linspace(0, 1, 10))
ax.plot_surface(XX, YY, ZZ, alpha=0.2, color='cyan')

# ------------------------------------------------------
# Configuraci√≥n de los ejes
# ------------------------------------------------------
ax.set_xlim([-0.5, 1.2])
ax.set_ylim([-0.5, 1.2])
ax.set_zlim([-0.5, 1.2])

# Etiquetas
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# ------------------------------------------------------
# T√≠tulo y leyenda
# ------------------------------------------------------
ax.set_title("Espacio ortonormal en $\\mathbb{R}^3$: vectores base y planos coordenados")
ax.legend()

# ------------------------------------------------------
# Mostrar gr√°fico
# ------------------------------------------------------
plt.show()
````
